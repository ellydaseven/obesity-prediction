### Converting the categorical variables to numerical ones via encoding

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()

# Converting categorical data 

data['Gender'] = encode.fit_transform(data['Gender'])
data['Family history'] = encode.fit_transform(data['Family history'])
data['Main meals per day'] = encode.fit_transform(data['Main meals per day'])
data['Water consumption'] = encode.fit_transform(data['Water consumption'])
data['Fast food consumption'] = encode.fit_transform(data['Fast food consumption'])
data['Fruit and vegetable consumption'] = encode.fit_transform(data['Fruit and vegetable consumption'])
data['Carbonated drinks consumption'] = encode.fit_transform(data['Carbonated drinks consumption'])
data['Foods between meals'] = encode.fit_transform(data['Foods between meals'])
data['Alcohol'] = encode.fit_transform(data['Alcohol'])
data['Sleep routine'] = encode.fit_transform(data['Sleep routine'])
data['Alcohol'] = encode.fit_transform(data['Alcohol'])
data['Tech device usage'] = encode.fit_transform(data['Tech device usage'])
data['Exercise'] = encode.fit_transform(data['Exercise'])
data['Sports activities'] = encode.fit_transform(data['Sports activities'])
data['Intense activity'] = encode.fit_transform(data['Intense activity'])
data['Obesity Status'] = encode.fit_transform(data['Obesity Status'])
```

```python
data = data.drop(['Gender_num'],axis=1)
```

### Using the SMOTE technique to oversample, to balance the dataset

```python
# Splitting the dataset

x = data.drop(['Obesity Status'],axis=1)
y = data['Obesity Status']
```

```python
# Import and use SMOTE 

from imblearn import over_sampling
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
x_bal,y_bal = oversample.fit_resample(x,y)
```

```python
x_bal.shape
```

```python
y_bal
```

```python
newData = pd.concat([x_bal,y_bal], axis=1, join="inner")
newData
```

```python
sns.catplot(x='Obesity Status', kind='count', data=newData)
```

```python
# Saving the dataset

newData.to_csv('DataBalanced.csv', index=False)
```
